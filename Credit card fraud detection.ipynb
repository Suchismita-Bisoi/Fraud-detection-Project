## Problem Statements

For many banks, retaining high profitable customers is the number one bussiness goal. Banking fraud, however, poses a significant threat to this goal for diffrent banks. In terms of substantial financial loses, trust and creditibility , this is a concernign issue to both banks and customers alike.

In the banking industry, credit card fraud detection using machine learning is not a only a trend but a necesity for them to put proactive monitoring and fraud prevention mechanism in place.Machine learning is helping these insttutions to reduce time-consuming manual reviews,costly chargebacks and fees as well as denials of legitimite transations.

In this project we will detect fraudulent credit card transaactions with the help of Machine learning models.We will analyse customer-level data that has been collected and analyed during a research colllbaration of worldline and the Machine Learning Group.

## Introduction

The data set includes credit card transactions made by European cardholders over a periods of two days in 2013. Out of total 2,84,807 transactions,492 were fraudlent. This dataset is a higly unbalalced., with the positive class accounting for 0.172% of the total transactions. The data set has aslso been modified with principal componnet analysis to maintain confidentiality. Apart from 'time' and 'amount', all the other fearture (V1,v2....V28) are principal components obtained using PCA. The feature 'time' contains the seconds elapesed between the first transaction in the data set and the subsequent transactions. The feature 'amount' is the transaction amount. The feature 'Class' represent s calss lebelling, and it take the value of 1 in cases of fraud and 0 in others.
